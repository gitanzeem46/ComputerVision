```python
# Create a zero-filled NumPy array with the specified shape (512, 512, 3)
# The shape should be passed as a tuple to np.zeros()
img = np.zeros((512,512,3))

# Import the alternative imshow function for Colab
from google.colab.patches import cv2_imshow

# Use the alternative imshow function to display the image
cv2_imshow(img)

# waitKey and destroyAllWindows are not needed with cv2_imshow in Colab
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(255,0,0),thickness=3)
cv2_imshow(img)

```

```python
cv2.rectangle(img,pt1=(100,100),pt2=(300,300),color=(255,0,0),thickness=-1)
cv2_imshow(img)
```

```python
cv2.circle(img,center=(100,100),radius=40,color=(0,0,255),thickness=3)
cv2_imshow(img)
```

```python
cv2.line(img,pt1=(0,0),pt2=(512,512),color=(0,255,0),thickness=3)
cv2_imshow(img)
```

```python
cv2.putText(img,org=(50,50), fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_AA,text="Hello",fontFace=cv2.FONT_ITALIC)
cv2_imshow(img)
```
